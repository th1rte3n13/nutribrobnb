```python
import torch
from PIL import Image
import matplotlib.pyplot as plt

# Load the pre-trained YOLOv5 model (YOLOv5s is the smallest model, you can try YOLOv5m, YOLOv5l for larger models)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Load an image (replace this with the path to your food image)
img = Image.open(r'D:\BITNBUILD2024\panipuri.jpg')

# Run the image through the model
results = model(img)

# Display the results
results.show()

# To print results as a Pandas DataFrame (including bounding boxes, labels, confidence)
print(results.pandas().xyxy[0])

# Save output image with detected food
results.save()  # It will save the output in a folder named 'runs'
```
